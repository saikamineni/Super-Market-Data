```python
#import

from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from array import *
import sys
import re

#load top-50 sales in each month

Location = 'C:\Users\saisree849\Desktop\Pandas\Vijay Project\jandata.csv'
df1 = pd.read_csv(Location)

Location = 'C:\Users\saisree849\Desktop\Pandas\Vijay Project\\febdata.csv'
df2 = pd.read_csv(Location)

Location = 'C:\Users\saisree849\Desktop\Pandas\Vijay Project\mardata.csv'
df3 = pd.read_csv(Location)

Location = 'C:\Users\saisree849\Desktop\Pandas\Vijay Project\\aprdata.csv'
df4 = pd.read_csv(Location)

Location = 'C:\Users\saisree849\Desktop\Pandas\Vijay Project\maydata.csv'
df5 = pd.read_csv(Location)
```

```python
#store all the product names into lists
a1 = df1['Description'].tolist()      
a2 = df2['Description'].tolist()
a3 = df3['Description'].tolist()
a4 = df4['Description'].tolist()
a5 = df5['Description'].tolist()

#This code stores all the products common in 5 months
common = []
for item1 in a1:
    for item2 in a2:
        if item1 == item2:
            for item3 in a3:
                if item1 == item3:
                    for item4 in a4:
                        if item1 == item4:
                            for item5 in a5:
                                if item1 == item5:
                                    common.append(item1)
```

```python
#stores quantities into list

#JANUARY
ref = []
for i in range(len(common)):
    for idx, record in df1['Description'].iteritems():
        if re.search(common[i], record):
            ref.append(df1.iloc[idx,1]) 
            
from pylab import *

# make a square figure and axes
figure(1, figsize=(10,10))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.


pie(ref, labels = common,
                autopct='%1.2f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('January Top-26', bbox={'facecolor':'0.8', 'pad':5})

show()


#FEBRUARY
ref = []
for i in range(len(common)):
    for idx, record in df2['Description'].iteritems():
        if re.search(common[i], record):
            ref.append(df2.iloc[idx,1]) 
            
from pylab import *

# make a square figure and axes
figure(1, figsize=(10,10))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.


pie(ref, labels = common,
                autopct='%1.2f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('February Top-26', bbox={'facecolor':'0.8', 'pad':5})

show()


#MARCH
ref = []
for i in range(len(common)):
    for idx, record in df3['Description'].iteritems():
        if re.search(common[i], record):
            ref.append(df3.iloc[idx,1]) 
            
from pylab import *

# make a square figure and axes
figure(1, figsize=(10,10))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.


pie(ref, labels = common,
                autopct='%1.2f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('March Top-26', bbox={'facecolor':'0.8', 'pad':5})

show()


#APRIL
ref = []
for i in range(len(common)):
    for idx, record in df4['Description'].iteritems():
        if re.search(common[i], record):
            ref.append(df4.iloc[idx,1])
            
from pylab import *

# make a square figure and axes
figure(1, figsize=(10,10))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.


pie(ref, labels = common,
                autopct='%1.2f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('April Top-26', bbox={'facecolor':'0.8', 'pad':5})

show()


#MAY
ref = []
for i in range(len(common)):
    for idx, record in df5['Description'].iteritems():
        if re.search(common[i], record):
            ref.append(df5.iloc[idx,1]) 
            
from pylab import *

# make a square figure and axes
figure(1, figsize=(10,10))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.


pie(ref, labels = common,
                autopct='%1.2f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('May Top-26', bbox={'facecolor':'0.8', 'pad':5})

show()

```
